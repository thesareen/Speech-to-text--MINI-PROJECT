# open the audio file using pydub
sound = AudioSegment.from_wav(path)
    #opening the audio segment from wav and given is the name of the file
    # split audio sound where silence is 700 miliseconds or more and get chunks
    #chunks mei sound ko divide krenge
    #split_on_silence is the function used
    #splits the sound with minimum silence length 500 and keep the silence for 1 sec
chunks = split_on_silence(sound,
        # experiment with this value for your target audio file
        min_silence_len = 500,
        # adjust this per requirement
        silence_thresh = sound.dBFS-14,
        # keep the silence for 1 second, adjustable as well
        keep_silence=500,
    )
folder_name = "audio-chunks"
    # create a directory to store the audio chunks
if not os.path.isdir(folder_name):
   os.mkdir(folder_name)#audio ko os mei save krenge and isdir returns true if folder_name is an existing directory
#agar not true means if folder_name dne then make a directory of folder_name
whole_text = ""#pass the text into the dir
    # process each chunk
for i, audio_chunk in enumerate(chunks, start=1):
  chunk_filename = os.path.join(folder_name, f"chunk{i}.wav")#each chunk transferred
  audio_chunk.export(chunk_filename, format="wav")#exporing the chunk
        # export audio chunk and save it in
        # the `folder_name` directory.

        # recognize the chunk
with sr.AudioFile(chunk_filename) as source:
  audio_listened = r.record(source)#the recorded audio will be recognized by google
            # try converting it to text

try:
  text = r.recognize_google(audio_listened)
except sr.UnknownValueError as e:
  print("Error:", str(e))

else:
  text = f"{text.capitalize()}. "
  print(chunk_filename, ":", text)
  whole_text += text

    # return the text for all chunks detected
return whole_text
